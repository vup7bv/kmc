**Q1.** This question is a case study for $k$ means clustering.

1. Load the `airbnb_hw.csv` data. Clean `Price` along with `Beds`, `Number of Reviews`, and `Review Scores Rating`.
2. Maxmin normalize the data and remove any `nan`'s (`KMeans` from `sklearn` doesn't accept `nan` input).
3. Use `sklearn`'s `KMeans` module to cluster the data by `Beds`, `Number of Reviews`, and `Review Scores Rating` for `k=6`.
4. Use `seaborn`'s `.pairplot()` to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.
5. Use `.groupby` and `.describe` to compute the average price for each cluster. Which clusters have the highest rental prices?
6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
````

````python
#loading data and cleaning column names

df = pd.read_csv('/data/airbnb_hw.csv')

df['price'] = df['Price']
df['beds'] = df['Beds']
df['n_reviews'] = df['Number Of Reviews']
df['score'] = df['Review Scores Rating']

# loc selects rows/columns in pandas
X = df.loc[:,['price','beds','n_reviews','score']]
# all rows, and the 4 columns

print(X.shape)
X.describe()
````

price doesn't even show up initially, and cleaning the column requires changing the string values to numeric, and removing the commas

````python
X['price'].value_counts()
X['price'] = X['price'].str.replace(',','') #removing commas from #s
X['price'] = pd.to_numeric(X['price'],errors='coerce')
X.describe()
````

````python
X['beds'] = X['beds'].fillna(1)
# it can be assumed that every listing has at least 1 bed
X.describe()
````

````python
X['n_reviews'] = pd.to_numeric(X['n_reviews'],errors='coerce')
X.describe()
````

````python
# figuring out how many values are missing
pd.crosstab(df['score'].isnull(), df['n_reviews']>0)
````

````python
X = X.dropna()
X.describe()
````

dropping the missing values, regardless of if they were due to 0 reviews or truly missing, does not seem to affect our group stats

````python
# maxmin normalizing data
def maxmin(x):
    u = (x-min(x))/(max(x)-min(x))
    return u
#removing nan values
X = X.dropna()

Z = X.drop('price',axis=1)
# created new df, without price column
Z = Z.apply(maxmin)
# df Z now has the normalized values for each column except price
````

````python
# using sklearn Kmeans to cluster data for k=6
from sklearn.cluster import KMeans # Import kmc

model = KMeans(n_clusters=6, max_iter=300, n_init = 10, random_state=0)
model = model.fit(Z) # Fit the emodel
Z['cluster'] = model.labels_
Z.describe()
````

````python
import seaborn as sns
# clustering carried out in many dimensions via scatterplots
sns.pairplot(data=Z,hue='cluster')
````

````python
# avg price for each cluster
X['cluster'] = model.labels_
X.loc[:,['price','cluster'] ].groupby('cluster').describe()
````

My descriptive stats likely differ a bit from the solutions since I wrote an extra line of code to remove the nan's from the X dataframe...

Cluster 4 is about 300/night (mean of 293) whereas the other clusters range from 130-150/night. These prices are reflected in the distribution amongst the 25/50/75 percentiles as well. Cluster 4 easily has the highest rental prices out of the group.

````python
# scree plot
k_bar = 15
k_grid = np.arange(1,k_bar+1)
SSE = np.zeros(k_bar)
for k in range(k_bar): # For indices from 0 to k_bar-1
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0) # Create a model for k
    model = model.fit(Z) # Fit the emodel
    SSE[k] = model.inertia_ # Record the sum of squared error for k in the SSE vector

sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot') # Create scree plot
````

````python
# k=3
model = KMeans(n_clusters=3, max_iter=300, n_init = 10, random_state=0) # Create a model for
model = model.fit(Z) # Fit the emodel
Z['cluster'] = model.labels_

sns.pairplot(data=Z,hue='cluster')

X['cluster'] = model.labels_
X.loc[:,['price','cluster'] ].groupby('cluster').describe()
````

Cluster 1 has high reviews on average, as does cluster 0. Clusters 0 and 2 have similarly low prices, but cluster 1 has the highest mean price.

    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df
